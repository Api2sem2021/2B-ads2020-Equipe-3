```python
#Importei a blibioteca Pandas
import pandas as pd
#Importei a blibioteca xlrd para ler dados e formatar informações de arquivos do Excel.
import xlrd
```

```python
#Declaração de variáveis globais
path_validados = "../3 - Output/2 - Validados/"
path_erros = "../3 - Output/1 - Erros/"
```

```python
#Localiza o documento
origem = (r"../../../1 - Tabelas Base/STG_MVT_CRD.xlsx")
#Para abrir em workbook (Pasta de trabalho, formato que encontrei que consiga modificar os dados )
documento_xlrd = xlrd.open_workbook(origem) 
celulas = documento_xlrd.sheet_by_index(0)
contador = 0
#Um for para ler as células da colunas 5 (Qual achei o erro)
for x in range (15):
    coluna_linha = (celulas.cell_value(x+1, 5))
    if coluna_linha == (""):
        contador = 1
        linha = x+1
print (f'Na linha {linha}, foi encontrado {contador} dado vazio.')
#Ler o documento em CSV, encontrar a codificação do idioma e um delimitador.
dados = pd.read_csv("../../../1 - Tabelas Base/STG_MVT_CRD.csv", encoding="UTF-8", delimiter=";")
```

```python
#Criei uma lista de números com quais usei para não serem apresentados na variavel linha_com_erro
lista = []
for y in range (16):
    if y != linha:
        lista.append(y)
del(lista[0])
linha_com_erro = pd.read_csv("../../../1 - Tabelas Base/STG_MVT_CRD.csv", encoding="UTF-8", delimiter=";", skiprows=lista)
linha_com_erro.to_excel(path_erros + "ERR_STG_MVT_CRD.xlsx",index=False)
```

```python
#Resultado da tabela com erro
linha_com_erro
```

```python
#Excluí a linha qual estava o erro e salvei em outro arquivo em XLSX
sem_erro = dados.drop(linha-1)
sem_erro.to_excel(path_validados + "MVT_CRD.xlsx",index=False)
```
