```python
import pandas as pd
low_memory=False
%matplotlib inline
```

```python
path = '../../../1 - Tabelas Base/'
filename = 'STG_PGT.xlsx'

path_erros = "../3 - Output/1 - Erros/"
path_validados = "../3 - Output/2 - Validados/"

#Mantém as informações dos pagamentos do Cadastro Positivo
df= pd.read_excel(path + filename)
```

```python
#Altera o nome das colunas
df=df.rename(columns={'ID_STG_PGT': 'ID_da_Stage', 'VLR_PGT_FAT': 'Valor_Pago', 'DAT_VCT': 'Data_Vencimento', 'COD_MDL': 'Modalidade_Pagamento', 'QTD_CLI_CAD_POS': 'CPF_Distintos', 'QTD_PGT': 'Qtd_Pagamentos', 
                      'ID_FNT_ITT': 'Fonte_Dados', 'DES_TIP_PSS': 'Tipo_Cliente', 'DAT_RSS_FNT_ITT': 'Data_Ultima_Alteracao', 'DAT_INC_DBO': 'Data_Inclusao',})
#Transforma a coluna em string e os outros três foi feita uma separação através de concatenação.
df['Data_Vencimento'] = df['Data_Vencimento'].apply(str) 
df['Dia'] = df['Data_Vencimento'].str.slice(0,2)
df['Mês'] = df['Data_Vencimento'].str.slice(2,4)
df['Ano'] = df['Data_Vencimento'].str.slice(4,)
#Junta três colunas em uma só
df["Data_Vencimento"] = df["Dia"]+ '-' + df["Mês"] + '-' + df["Ano"]
#Excluí colunas
df=df.drop(columns=['Dia','Mês','Ano'])
df
```

```python
df[['ID_da_Stage', 'Valor_Pago',  'Data_Vencimento',  'Modalidade_Pagamento','CPF_Distintos','Qtd_Pagamentos', 
    'Fonte_Dados','Tipo_Cliente','Data_Ultima_Alteracao', 'Data_Inclusao']].isnull().sum()

#O comando isnull permite confirmar em todas as colunas, se em alguma célula possuí dado nulo ("Nan").
#No output é possível perceber que não existe esse caso em nenhuma das células
```

```python
df1= df[df.CPF_Distintos > df.Qtd_Pagamentos]
df1
#O comando confirma se não há nenhum dado na coluna E que seja maior que o dado da coluna F, que seria um erro.
#A planilha não possuí nenhum desses casos, portanto está livre desse tipo de erro
```

```python
df = df[~df.ID_da_Stage.isin(df1)]
```

```python
##Exporta as df
df.to_excel(path_validados + 'PGT.xlsx', index=False)
df.to_excel(path_erros + 'ERR_STG_PGT.xlsx', index=False)
```
