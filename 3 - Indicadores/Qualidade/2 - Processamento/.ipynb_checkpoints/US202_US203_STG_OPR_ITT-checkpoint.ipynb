```python
#Parametrizações iniciais - import das bibliotecas.

import pandas as pd
import openpyxl
import xlrd

low_memory=False
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
```

```python
path_erros = "../3 - Output/1 - Erros/"
path_validados = "../3 - Output/2 - Validados/"
```

```python
# Geração do DATAFRAME completo do arquivo.

df_STG_OPR_ITT = pd.read_excel(r'../../../1 - Tabelas Base/STG_OPR_ITT.xlsx', delimiter=';', error_bad_lines=False, encoding='UTF-8')
```

```python
#Apresentação de uma amostra dos dados. 

df_STG_OPR_ITT.sample(10)
```

```python
#Script gera um DATAFRAME contendo as linhas que não apresentam divergência da condição:
#O campo VLR_CTRD_CSC só poderá ter valores NUMÉRICOS se seu respectivo ID_MDL for igual a C01

df_VLR_CTRD_CSC_Corretos = df_STG_OPR_ITT.query('(ID_MDL == "C01") and (VLR_CTRD_CSC > 0)')
```

```python
#Impressão do DATAFRAME

print(f'Quantidade de registro(s) Correto(s): {df_VLR_CTRD_CSC_Corretos.shape[0]}.')
df_VLR_CTRD_CSC_Corretos
```

```python
#Script gera um DATAFRAME contendo as linhas que apresentam divergência da condição:
#O campo VLR_CTRD_CSC só poderá ter valores NUMÉRICOS se seu respectivo ID_MDL for igual a C01

df_VLR_CTRD_CSC_Incorretos = df_STG_OPR_ITT.query('((ID_MDL != "C01") and (VLR_CTRD_CSC > 0)) or ((ID_MDL == "C01") and (VLR_CTRD_CSC == ""))')
```

```python
#Impressão do DATAFRAME INCORRETOS

print(f'Quantidade de registro(s) Incorreto(s): {df_VLR_CTRD_CSC_Incorretos.shape[0]}.')
df_VLR_CTRD_CSC_Incorretos
```

```python
#Consolidação das Informações Obitidas:

totalRegistros = df_STG_OPR_ITT.shape[0]
totalCorretos = df_VLR_CTRD_CSC_Corretos.shape[0]
totalIncorretos = df_VLR_CTRD_CSC_Incorretos.shape[0]
consolidacao = {"Quantidade": [df_STG_OPR_ITT.shape[0], df_VLR_CTRD_CSC_Corretos.shape[0], df_VLR_CTRD_CSC_Incorretos.shape[0]], 
                "%": [f'100',f'{((df_VLR_CTRD_CSC_Corretos.shape[0]*100)/df_STG_OPR_ITT.shape[0])}', f'{((df_VLR_CTRD_CSC_Incorretos.shape[0]*100)/df_STG_OPR_ITT.shape[0])}']}
```

```python
#Geração do DATAFRAME com os dados de CONSOLIDAÇÃO.

df_consolidacao = pd.DataFrame(consolidacao, columns=['Quantidade', '%'], index=['Total Registros', 'Total Corretos', 'Total Incorretos'])
```

```python
#Impressão do DATAFRAME CONSOLIDAÇÃO.

df_consolidacao
```

```python
## Exportar df para xlsx em ordem crescente e sem index

df_VLR_CTRD_CSC_Corretos.sort_values('ID_STG_OPR_ITT').to_excel(path_erros + 'ERR_STG_OPR_ITT.xlsx', index = False)
df_VLR_CTRD_CSC_Incorretos.sort_values('ID_STG_OPR_ITT').to_excel(path_validados + 'OPR_ITT.xlsx', index= False)
```
