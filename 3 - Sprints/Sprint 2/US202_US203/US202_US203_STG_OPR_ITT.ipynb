```python
# Geração do DATAFRAME completo do arquivo.

df_STG_OPR_ITT = pd.read_csv(r'.\data\STG_OPR_ITT.csv', delimiter =',', encoding='iso-8859-1')
```

```python
#Script gera um DATAFRAME contendo as linhas que não apresentam divergência da condição:
#O campo VLR_CTRD_CSC só poderá ter valores NUMÉRICOS se seu respectivo ID_MDL for igual a C01

df_VLR_CTRD_CSC_Corretos = df_STG_OPR_ITT.query('(ID_MDL == "C01") and (VLR_CTRD_CSC > 0)')
```

```python
#Impressão do DATAFRAME

print(f'Quantidade de registro(s) Correto(s): {df_VLR_CTRD_CSC_Corretos.shape[0]}.')
df_VLR_CTRD_CSC_Corretos
```

```python
#Script gera um DATAFRAME contendo as linhas que apresentam divergência da condição:
#O campo VLR_CTRD_CSC só poderá ter valores NUMÉRICOS se seu respectivo ID_MDL for igual a C01

df_VLR_CTRD_CSC_Incorretos = df_STG_OPR_ITT.query('((ID_MDL != "C01") and (VLR_CTRD_CSC > 0)) or ((ID_MDL == "C01") and (VLR_CTRD_CSC == ""))')
```

```python
#Impressão do DATAFRAME INCORRETOS

print(f'Quantidade de registro(s) Incorreto(s): {df_VLR_CTRD_CSC_Incorretos.shape[0]}.')
df_VLR_CTRD_CSC_Incorretos
```

```python
#Consolidação das Informações Obitidas.

totalRegistros = df_STG_OPR_ITT.shape[0]
totalCorretos = df_VLR_CTRD_CSC_Corretos.shape[0]
totalIncorretos = df_VLR_CTRD_CSC_Incorretos.shape[0]
consolidacao = {"VLR_CTRD_CSC & ID_MDL": [df_STG_OPR_ITT.shape[0], df_VLR_CTRD_CSC_Corretos.shape[0], df_VLR_CTRD_CSC_Incorretos.shape[0]], 
                "%": [f'100',f'{((df_VLR_CTRD_CSC_Corretos.shape[0]*100)/df_STG_OPR_ITT.shape[0])}', ((df_VLR_CTRD_CSC_Incorretos.shape[0]*100)/df_STG_OPR_ITT.shape[0])]}
```

```python
#Geração do DATAFRAME com os dados de CONSOLIDAÇÃO.

df_consolidacao = pd.DataFrame(consolidacao, columns=['VLR_CTRD_CSC & ID_MDL', '%'], index=['Total Registros', 'Total Corretos', 'Total Incorretos'])
```
